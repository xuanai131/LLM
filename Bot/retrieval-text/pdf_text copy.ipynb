```python
import PyPDF2

def is_valid_pdf(file_path):
    try:
        with open(file_path, 'rb') as file:
            reader = PyPDF2.PdfFileReader(file)
            _ = reader.numPages  # Try accessing number of pages
            return True
    except PyPDF2.utils.PdfReadError:
        return False

file_path = "/home/xuanai/html/Bot/Knowledge/Books/PDF/Nghệ thuật - Ẩm thực/skn010475_3278.pdf"

if is_valid_pdf(file_path):
    print("Valid PDF file.")
else:
    print("Invalid PDF file.")
```

```python
from langchain_community.vectorstores import Chroma

import sys
sys.path.append("../")
import API_keys
from langchain_openai import OpenAIEmbeddings
embedding=OpenAIEmbeddings(chunk_size=1)
from methods import *


# Load book_info database
# SelfKnowledge =  DATABASE(db_path='robot_knowledge_2', embedding=embedding, parent_path="parents_2")
SelfKnowledge =  DATABASE(db_path='robot_knowledge_3', embedding=embedding, parent_path="parents_3")



##### CREATE AGENT FUNCTION
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.agents import AgentExecutor, create_openai_tools_agent
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-3.5-turbo-0125")
def create_agent(llm: ChatOpenAI, tools: list, system_prompt: str):
    # Each worker node will be given a name and some tools.
    prompt = ChatPromptTemplate.from_messages(
        [
            (
                "system",
                system_prompt,
            ),
            MessagesPlaceholder(variable_name="messages"),
            MessagesPlaceholder(variable_name="agent_scratchpad"),
        ]
    )
    agent = create_openai_tools_agent(llm, tools, prompt)
    executor = AgentExecutor(agent=agent, tools=tools)
    return executor

# import Tools
from Prompt import *
from langchain.tools.retriever import create_retriever_tool
book_search_tool = create_retriever_tool(
        SelfKnowledge.retriever,
        "book_researcher",
        "tìm kiếm và trả lời các thông tin về sách trong database của thư viện cho người dùng",
    )

book_research_agent = create_agent(llm, [book_search_tool], "Bạn rất hữu dụng trong việc trả lời các câu hỏi về sách cho người dùng. Lưu ý: câu trả lời phải bằng tiếng việt và không quá dài dòng.")
```
